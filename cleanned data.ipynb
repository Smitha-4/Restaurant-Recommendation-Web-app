```python
#standard libraries
import pandas as pd
import numpy as np
from cleantext import clean
```

```python
file1=pd.read_csv('../dataset files/zomato.csv')
file2=pd.read_csv('../dataset files/Swiggy Bangalore.csv')
```

```python
file2=file2.rename(columns={'Restaurant Name':'name','Category':'rest_type','Rating':'rate','Cost for Two (in Rupees)':'approx_cost(for two people)',
                     'Area':'location','URL':'url'})
```

```python
df = file1.append(file2, ignore_index=True)
```

```python
df.reset_index(drop=True)
df
```

```python
df=df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
df
```

```python
df=df.drop(['Offer'],axis=1)
df=df.drop(['city'],axis=1)
df=df.drop(['menu_item'],axis=1)
```

I am not going to drop the columns except for offer column as they will be come in handy in the final product. The offer is an inclusive thing which the restaurants decide depending on the season. I will clean rate column, votes column, cost column as they are numerical columns. 

**Another important thing is to clean categorical columns such as address, online order, book table, rest type, dishes liked, cuisines, review list, menu item. I am going to use Text Summarize and Tensorflow Vector Detection**

```python
df['address'].fillna('Sorry! The adress you are looking for is not found.', inplace=True )
df['online_order'].fillna('No', inplace=True)
df['book_table'].fillna('No',inplace=True)
df['phone'].fillna('Sorry! Phone Number Not Available.', inplace=True)
df['votes'].fillna(0,inplace=True)
df['dish_liked'].fillna('This data is not available', inplace=True)
df['cuisines'].fillna(df['rest_type'],inplace=True)
df['reviews_list'].fillna('The Review for this restaurant is not avialable.',inplace=True)
df['type'].fillna('unknown',inplace=True)
```

```python
def convert_rate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(convert_rate)
df.fillna(0, inplace=True)
df
```

```python
def make_integer(value):
    value=str(value)
    if ',' in value:
        value= value.replace(',','')
        return int(value)
    else:
        return int(value)
df['cost']=df['cost'].apply(make_integer)
df['cost'].unique()
```

```python
dish_liked=df['dish_liked'].value_counts(ascending=False)
dish_liked
dish_0=dish_liked[dish_liked==28078]
dish_0
```

```python
def make_other_dish(value):
    if(value in dish_0):
        return 'not rated item'
    else:
        return value
df['dish_liked']=df['dish_liked'].apply(make_other_dish)
df['dish_liked'].value_counts()
```

```python
dish_50=dish_liked[dish_liked<=50]
dish_50
```

```python
for i, element in enumerate(df['reviews_list']):
    df.loc[i,'reviews_list']=clean(df.loc[i,'reviews_list'],fix_unicode=True,to_ascii=True,                  
                                lower=True,                     
                                no_line_breaks=True,           
                                no_urls=True,                 
                                no_emails=True,                
                                no_phone_numbers=True,         
                                no_numbers=True,               
                                no_digits=True,                
                                no_currency_symbols=True,      
                                no_punct=True,                 
                                replace_with_punct="",          
                                replace_with_url="",
                                replace_with_email="",
                                replace_with_phone_number="",
                                replace_with_number="",
                                replace_with_digit="",
                                replace_with_currency_symbol="",
                                lang="en")
df['reviews_list']=df['reviews_list'].str.replace('rated','')
df['reviews_list']=df['reviews_list'].str.replace('\n','')
```

```python
df.to_csv('cleanned_data.csv',index=False)
```
